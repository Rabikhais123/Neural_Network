```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Pandas dataset/breast-cancer.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
```

```python
X=df.iloc[:,2:]
y=df.iloc[:,1]
```

```python
X
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Scaling
from sklearn.preprocessing import MinMaxScaler
me=MinMaxScaler()
X_train=me.fit_transform(X_train)
X_test=me.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann=tf.keras.models.Sequential()
```

```python
#add input Layer
ann.add(Dense(units=30,activation='relu'))               #dense layer is called fully connected layer
```

```python
#add first hidden layer
ann.add(Dense(units=8,activation='relu'))
```

```python
ann.add(Dense(units=6,activation='relu'))
```

```python
ann.add(Dense(units=4,activation='relu'))
```

```python
#output layer
ann.add(Dense(units=1,activation='sigmoid'))
```

```python
#weight adjust optimizer
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
ann.summary()
```

```python
y_pred=ann.predict(X_test)
y_pred
```

```python
y_pred=(y_pred>0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
print(accuracy_score(y_test,y_pred)*100)
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
